```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url_data=r'https://raw.githubusercontent.com/astroclubiitk/computational-astrophysics/main/Week-1/Assignment/Astrosat_Catalog.csv'
df=pd.read_csv(url_data)
df.head()
```

```python
df.loc[df['ra']>180,'ra1']=(df['ra']-360)*np.pi/180
df.loc[df['ra']<=180,'ra1']=(df['ra']*np.pi/180)
df.insert(7,'dec2',df['dec']*np.pi/180)
h=df[df.Final_Type=='HMXB']
l=df[df.Final_Type=='LMXB']
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,sharex=True)
ax1.plot(h.ra,h.dec,'k-',label='HMXB')
ax1.set_title('HMXB(High Mass X-ray Binaries)')
ax1.set_xlabel('ra')
ax1.set_ylabel('dec')
ax1.legend(loc=4)            
#plotting axis1
ax2.plot(l.ra,l.dec,'b',label='LMXB')
ax2.set_title('(Low Mass X-Ray Binaries)')
ax2.set_xlabel('ra')
ax2.set_ylabel('dec')
plt.legend(loc=4)
plt.tight_layout()
plt.show()
#plotting axis2
```
